### Ingestion Cell

```python
import pandas as pd
from io import BytesIO
from db import SupabaseHandler
from embed import embed_documents

def ingest_csv_to_supabase(file_path: str, table_name: str = "book_as_concept_card"):
    
    print(f"Reading file: {file_path}...")

    with open(file_path, "rb") as f:
        binary_data = f.read()
    df = pd.read_csv(BytesIO(binary_data))

    text_to_embed = (df['concept_name'] + ": " + df['scenario']).tolist()
    embeddings = embed_documents(text_to_embed)

    df['embedding'] = embeddings

    print(f"Ingesting to Supabase table: {table_name}...")

    db = SupabaseHandler()
    records = df.to_dict(orient='records')

    try:

        result = db.insert(table_name, records)
        print(f"Successfully ingested {len(result)} rows.")

    except Exception as e:
        print(f"Error during ingestion: {e}")

if __name__ == "__main__":
    # CSV_FILE = "Annotation of Books and Analysis Conversations - book_as_concept_card.csv"
    # ingest_csv_to_supabase(CSV_FILE)
```
